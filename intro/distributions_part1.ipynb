### Random Variable 
* A Variable whose probability outcomes of a random phenomenon 
* A Discrete random variable is one which may take only a  countable number of discrete values and thus can be quantified 
    * eg: A number which comes up when you roll a fair dice 
    * The probability distribution of a discrete random variable is a list of probabilities associated with each of its possible values (also know as a probability function or a probability mass function). 
    * mathematical sense: random variable X may take k different values with probabilities X = x<sub>i</sub> defined to be P(X=x<sub>i</sub>) = p<sub>i</sub> then thee probabilities p<sub>i</sub> must follow the following.
        * 0<p<sub>i</sub><1
        * p<sub>1</sub>+p<sub>2</sub>+...+p<sub>k</sub>=1
    * examples of discrete probability are Bernoulli distribution, Binomial distribution, Possion distribution. 
* A continous random variable is one which takes an infinite number of possible values.
    * eg: height of students in a class 
    * Since the continous random variable is defined over an interval of values, it is represented by area under the curve (or the integral). 
    * The probability distribution of continous random variables are functions that take on continous values. The probanility of observing a single value is 0 since the number of values which may be assumed by random variable may be infinite. 
    * The curve which represents p(x) must satisty the following
        * The curve has no negative values (p(x) > 0 for all x)
        * The total area under the curve = 1 
* There are other types of distribution that pops up in litrature one should know about called cumulative distribution function.
    * All random variable X discrete and continous have a cumulative distribution function. 
    * It is a function giving the probability that the random variable X is less than or equal to x, for every value of x.
    * For a discrete random variable the distribution function is found by summing up the probabilities.

#### 1. Uniform distribution

The probability distribution function of a continous uniform distribution is 

![Fig 1: Uniform Distribution Function](https://res.cloudinary.com/dyd911kmh/image/upload/f_auto,q_auto:best/v1539784816/output_5_0_f453qp.png "Fig 1: Uniform Distribution Function")

* Since any interval of numbers of equal width has an equal probability of being observed, the curve describing the distribution is a rectangle, with constant height across the interval and zero height elsewhere. 
* Since the area under the interva is supposed to be 1, the lenght of the interval determines the height of the curve

```python
# import the uniform distribution 
from scipy.stats import uniform 
"""
The uniform function generates a uniform continous variable between specified interval via its loc and sacle arguments.
This distribution is constant between loc and loc+scale. Thee size argument describes the number of random variates. 
The size argument describes the number of random variates. 
"""
n = 10000 
start = 10 
width = 20 
data_uniform = uniform.rvs(size=n, loc=start, scale=width)
```

```python
ax = sns.distplot(data_uniform, bins=100, kde=True, color="skyblue", hist_kws={"linewidth":15, "alpha":1})
ax.set(xlabel="Uniform_Distribution", ylabel="Frequency")
```
